```python
import pandas as pd

jan = "Resources/citibike_jan.csv"
feb = "Resources/citibike_feb.csv"
march = "Resources/citibike_march.csv"
april = "Resources/citibike_april.csv"
may = "Resources/citibike_may.csv"
june = "Resources/citibike_june.csv"
july = "Resources/citibike_july.csv"
aug = "Resources/citibike_august.csv"
sept = "Resources/citibike_september.csv"
octo = "Resources/citibike_october.csv"
nov = "Resources/citibike_november.csv"
dec = "Resources/citibike_december.csv"

jan_df = pd.read_csv(jan)
feb_df = pd.read_csv(feb)
march_df = pd.read_csv(march)
april_df = pd.read_csv(april)
may_df = pd.read_csv(may)
june_df = pd.read_csv(june)
july_df = pd.read_csv(july)
aug_df = pd.read_csv(aug)
sept_df = pd.read_csv(sept)
oct_df = pd.read_csv(octo)
nov_df = pd.read_csv(nov)
dec_df = pd.read_csv(dec)
```

```python
janlength = len(jan_df)
feblength = len(feb_df)
marlength = len(march_df)
aprlength = len(april_df)
maylength = len(may_df)
junelength = len(june_df)
julylength = len(july_df)
auglength = len(aug_df)
seplength = len(sept_df)
octlength = len(oct_df)

            
print(f" January: {janlength}")
print(f" February: {feblength}")
print(f" March: {marlength}")
print(f" April: {aprlength}")
print(f" May: {maylength}")
print(f" June: {junelength}")
print(f" July: {julylength}")
print(f" August: {auglength}")
print(f" September: {seplength}")
print(f" October: {octlength}")
```

```python
jan_df_limit = jan_df.sample(n=30000, random_state=1, )
feb_df_limit = feb_df.sample(n=30000, random_state=1)
mar_df_limit = march_df.sample(n=30000, random_state=1)
april_df_limit = april_df.sample(n=30000, random_state=1)
may_df_limit = may_df.sample(n=30000, random_state=1)
june_df_limit = june_df.sample(n=30000, random_state=1)
july_df_limit = july_df.sample(n=30000, random_state=1)
aug_df_limit = aug_df.sample(n=30000, random_state=1)
sept_df_limit = sept_df.sample(n=30000, random_state=1)
oct_df_limit = oct_df.sample(n=30000, random_state=1)
nov_df_limit = nov_df.sample(n=30000, random_state=1)
dec_df_limit = dec_df.sample(n=30000, random_state=1)
```

```python
####

citbike_months = [jan_df_limit, feb_df_limit, mar_df_limit, april_df_limit, may_df_limit, june_df_limit, july_df_limit, aug_df_limit, sept_df_limit, oct_df_limit, nov_df_limit, dec_df_limit]



```

```python
df = pd.concat(citbike_months)
original_length = len(df)
print(original_length)
```

```python
# Remove all rows where trip duration is less than 5 minutes
df = df[df['tripduration'] >= 300 ] 

# Remove all rows where trip duration is more than 12 hours
df = df[df['tripduration'] <= 43200 ] 

outliers_removed_length = len(df)
print(outliers_removed_length)
```

```python
#Add tripduration(min)

df['tripduration(minutes)'] = df['tripduration'] / 60

df['tripduration(hours)'] = df['tripduration(minutes)'] / 60
```
